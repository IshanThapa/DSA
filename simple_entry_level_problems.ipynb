```python
def generate_fibonacci(n):
    # Initialize the first two terms
    fib_series = [0, 1]

    # Handle cases where n is 1 or 2
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return fib_series

    # Calculate terms from the 3rd index up to n
    while len(fib_series) < n:
        next_value = fib_series[-1] + fib_series[-2]
        fib_series.append(next_value)

    return fib_series

# Example usage: Generate the first 10 numbers
print(generate_fibonacci(10))
```
